````python
sequences = [
    ("seq1", "MKTAYIAKQRQISFVKSHFSRQDILDLI"),
    ("seq2", "MKLAYIAKQRQISFVKSHFSRQDILDLI"),
    ("seq3", "MKTAYIAKQRQISFVKSHFSRQDILDLV"),
    ("seq4", "MKTAYIAKQRQISFVKSHFSRQDILDLI"),
    ("seq5", "MKTAYIAKIAKQRQISFVKSHFSRQDILDLI"),
    ("seq6", "MKTAYIAKIAKQRQISFVKSHFSRQDILDLI"),
    ("seq7", "MKTAYIAKIAKQRQISFVSHFSRQDILDLI"),
    ("seq8", "MKTAYIAKIAKQRQISFVSHFSRQDILDLI"),
    ("seq9", "MKTAYIAKIAKQRQISFVSHFSRQDILDLI"),
    ("seq10", "MKTAYIAKQRQISFVKSHFSRQDILDLI"),
    ("seq11", "MKTAYIAKQRQISFVSHFSRQDILDLI"),
]

# Create a default configuration
# The attributes of the configuration mirror
# the command line interface
config = Configuration()

# Adapt the config as needed
# For example, set a custom number of models
config.training.num_model = 3

# Load a dataset and align the sequences
with SequenceDataset(sequences=sequences) as data:
    alignment_model = align(data, config)

alignment_model.to_string()
````
